今回は、[小売物価統計調査 小売物価統計調査（動向編）主要品目の都市別小売価格－都道府県庁所在市及び人口15万以上の市(2000年1月～) | 統計表・グラフ表示 | 政府統計の総合窓口](https://www.e-stat.go.jp/dbview?sid=0003421913)のデータを加工して分析を行います。

出典：政府統計の総合窓口(e-Stat)（https://www.e-stat.go.jp/　）

```python
import pandas as pd

DATA_PATH="../datasets/FEH_00200571_241119204010.csv"
df = pd.read_csv(DATA_PATH, encoding="shift_jis")

df.head()
```

```python
def _convert_to_float(x):
    try:
        return float(x.replace(",", ""))
    except:
        return None

df["value"] = df["value"].apply(
    _convert_to_float
)
```

```python
df_sapporo = df[df["地域"] == "札幌市"]
df_sapporo["value"] = df_sapporo["value"].astype(float)
```

```python
from prophet import Prophet
```

```python
# df has "時間軸"　as YYYY年MM月 and "value" as float

df_sapporo = df_sapporo.rename(
    columns={
        "時間軸（月）": "ds",
        "value": "y"
    }
)

# convert YYYY年MM月 to YYYY-MM-01
# 1月~9月については、先頭に0を追加する
df_sapporo["ds"] = df_sapporo["ds"].apply(
    lambda x: x.replace("年", "-").replace("月", "-01") 
    if len(x) == 8
    else x.replace("年", "-0").replace("月", "-01")
)

model = Prophet(
    seasonality_mode="multiplicative",
    yearly_seasonality=True,
    weekly_seasonality=False,
    daily_seasonality=False
)

model.fit(df_sapporo)

figure = model.plot(model.predict(df_sapporo))
figure_component = model.plot_components(model.predict(df_sapporo))
```

```python
# 年と月を抽出
df_sapporo["year"] = pd.to_datetime(df_sapporo["ds"]).dt.year
df_sapporo["month"] = pd.to_datetime(df_sapporo["ds"]).dt.month

# ピボットテーブルを作成
pivot_df = df_sapporo.pivot(index="year", columns="month", values="y")

highest_month = pivot_df.idxmax(axis=1)
lowest_month = pivot_df.idxmin(axis=1)

pivot_df["最高月"] = highest_month
pivot_df["最低月"] = lowest_month

pd.set_option('display.max_columns', 100)

# 結果を表示
pivot_df
```

```python
import matplotlib.pyplot as plt
import japanize_matplotlib

# drop the highest and lowest month columns
try:
    pivot_df = pivot_df.drop(columns=["最高月", "最低月"])
except:
    pass

# Plot each year's data as a line
plt.figure(figsize=(10, 6))
for year in pivot_df.index:
    plt.plot(pivot_df.columns, pivot_df.loc[year], label=f"{year}年", marker='o')

# Customize the plot
plt.title("各年ごとの値段（月別）")
plt.xlabel("月")
plt.ylabel("値段")
plt.xticks(range(1, 13))  # Ensure all months are displayed
plt.legend(title="年")
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```
